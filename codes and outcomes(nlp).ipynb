```python
import nltk
import re
from nltk.corpus import stopwords
import string
from nltk.corpus import sentence_polarity
import random
```

```python
f = open("finefoods.txt",encoding = "ISO-8859-1")
raw=f.read()
```

```python
pattern='(product\/productId: )([0-9A-Z]+)\s(review\/userId: )([0-9A-Z]+)\s(review\/profileName: )([A-Za-z1-9,.?!\'\";:@#$%^&*()\-_=+`~[{\]}\\|\/<>\s]+)(review\/helpfulness: )(\d\/\d)\s(review\/score: )(\d.\d)\s(review\/time: )(\d{10})\s(review\/summary: )([A-Za-z1-9,.?!\'\";:@#$%^&*()\-_=+`~[{\]}\\|\/<>\s]+)\s(review\/text: )([A-Za-z1-9,.?!\'\";:@#$%^&*()\-_=+`~[{\]}\\|\/<>\s]+)\s(product\/productId: )'


```

```python
matches = re.findall(pattern,raw)
```

```python
f.close()
```

```python
print(matches[0])
```

```python
len(matches)
```

```python
productID = []
userID = []
review_helpfulness = []
review_score = []
review_summary = []
review_text = []

for m in matches:
    productID.append(m[1])
    userID.append(m[3])
    review_helpfulness.append(m[7])
    review_score.append(m[9])
    review_summary.append(m[13])
    review_text.append(m[15])
```

```python
import pandas as pd
```

```python
d = {
    'Product ID': productID,
    'User ID': userID,
    'Review Helpfulness': review_helpfulness,
    'Review Score': review_score,
    'Review Summary': review_summary,
    'Review Text': review_text
}
```

```python
raw_data = pd.DataFrame(data = d)
```

```python
raw_data
```

```python
raw_data['Review Score'] = [float(i) for i in raw_data['Review Score']]
```

```python
User_Filter = raw_data.groupby(['User ID']).agg({'Review Text' : 'count',
                                  'Review Score': 'mean'})
```

```python
User_Filter
```

```python
set(User_Filter['Review Text'])
# The number of reviews of the customer ranges from 1 to 180
```

```python
# So we select the customers who gives more than 20 reviews. 
```

```python
more_than_20_reviews_users_agg_df = User_Filter.loc[(User_Filter['Review Text'] >20)]
```
